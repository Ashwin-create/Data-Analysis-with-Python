# Data Cleaning

```python
# next step is Data Cleaning
#drop unrelated blank columns
dataframe.drop(['Status', 'unnamed1'], axis=1, inplace=True)
#axis=1 means that entire vertical row, inplace=True means save it
```

```python
#check for null values
pd.isnull(dataframe).sum()
```

```python
#to drop null values
dataframe.dropna(inplace=True)
```

```python
dataframe.shape
```

```python
#to change data type of a col
dataframe['Amount'] = dataframe['Amount'].astype('int')
```

```python
dataframe['Amount'].dtypes
```

```python
dataframe.columns
```

```python
#rename column
dataframe.rename(columns= {'Marital_Status': 'Married'})
```

# Exploratory Data Analysis (EDA)

```python
dataframe.columns
```

```python
#Gender
gender_graph = sns.countplot(x = 'Gender', data = dataframe)

for bars in gender_graph.containers:
    gender_graph.bar_label(bars)
```

```python
dataframe.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = dataframe.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

ax = sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)

for bars in ax.containers:
    ax.bar_label(bars)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females is greater than men.*

```python
#Age
ax = sns.countplot(data = dataframe, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount vs Age Group
sales_age = dataframe.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(data = sales_age, x='Age Group', y='Amount')
```

*From above graphs we can see that most of the buyers are from 26-35 age group and are females.*

```python
#State
#total number of orders from top 10 states

sales_state = dataframe.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

#to set the size of the chart so that state names do not overlap
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data = sales_state, x='State', y='Orders')
```

```python
#total amount/sales from top 10 states
sales_state = dataframe.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data = sales_state, x='State', y='Amount')
```

*From above graphs we can see that most orders & total sales/amount are from UP, Maharashtra and Karnataka respectively.

```python
#Marital Status

ax = sns.countplot(data = dataframe, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(3,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_married = dataframe.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(3,5)})
sns.barplot(data = sales_married, x='Marital_Status', y='Amount', hue = 'Gender')
```

*From above graphs we can see that most of the buyers are married women and they have high purchasing power

```python
#Occupation
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = dataframe, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ = dataframe.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_occ, x = 'Occupation', y = 'Amount')
```

*From above graphs we can see that most buyers are from IT Sector, Healthcare, Aviation.

```python
#Product Category

sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = dataframe, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_pc = dataframe.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_pc, x = 'Product_Category', y = 'Amount')
```

*We can see the most products are sold from Clothing & Apparel, Food and Electronics & Gadgets
*Top 10 product categories sales as per amount is shown in the second graph.

```python
sales_pi = dataframe.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
#head(10) = nlargest(10). means the same.
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_pi, x = 'Product_ID', y = 'Orders')
```

```python
# Another way to find top 10 products 

fig1, ax1 = plt.subplots(figsize=(12,7))
dataframe.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

# Conclusion

**Married women from age group 26-35 years from UP, Maharashtra and Karnataka working in the IT sector, Healthcare and Aviation are more likely to buy products from Food, Clothing & Apparel and Electronics & Gadgets category.
